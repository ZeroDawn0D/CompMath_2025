```python
import numpy as np
import matplotlib.pyplot as plt
from ipynb.fs.full.Thomas_Algo import ThomasAlgorithm,create_tridiagonal
```

```python
def create_matrix(y,x,**kwargs):
    args  = kwargs["args"]
    alpha = args[0]
    
    Nx = len(x)
    dx = x[1]-x[0]
    a  = np.zeros(len(x))
    b  = np.zeros(len(x))
    c  = np.zeros(len(x))
    d  = np.zeros(len(x))
    
    for n in range(1,Nx-1,1):
        a[n] = -alpha
        b[n] = (1+2*alpha)
        c[n] = -alpha
        d[n] = y[n] #dx**2*f(x[n]) #y[n]

    return a,b,c,d
```

```python
def Crank_Nicholson(y,x,**kwargs):
    args  = kwargs["args"]
    alpha = args[0]
    
    Nx = len(x)
    dx = x[1]-x[0]
    a  = np.zeros(len(x))
    b  = np.zeros(len(x))
    c  = np.zeros(len(x))
    d  = np.zeros(len(x))
    
    for n in range(1,Nx-1,1):
        a[n] = -0.5*alpha
        b[n] = (1+alpha)
        c[n] = -0.5*alpha
        d[n] = y[n]*(1-alpha) + 0.5*alpha*y[n+1]+0.5*alpha*y[n-1]

    return a,b,c,d
```

```python
Nx     = 101
xmax   = 10
xmin   = 0
x      = np.linspace(0,xmax-xmin,Nx)
t      = np.linspace(0,1,1001)
sol    = np.zeros((len(t),len(x)))
sol[0,:] = np.sin(3*np.pi*x/(xmax-xmin))
sol_CN = np.zeros((len(t),len(x)))
sol_CN[0,:] = np.sin(3*np.pi*x/(xmax-xmin))
sol_exp =  np.zeros((len(t),len(x)))
sol_exp[0,:] = np.sin(3*np.pi*x/(xmax-xmin))
D        = 1.0
dx       = x[1]-x[0]
dt       = t[1]-t[0]
alpha    = D*dt/(dx**2)
print(len(x))
```

```python
for time in range(1,len(t),1): 
    sol[time] = ThomasAlgorithm(create_matrix,sol[time-1],x,args=(alpha,),Left="Dirichlet",Right="Dirichlet")
    #sol_CN[time] = ThomasAlgorithm(create_matrix,sol_CN[time-1],x,args=(alpha,),Left="Dirichlet",Right="Dirichlet")
    sol_CN[time] = ThomasAlgorithm(Crank_Nicholson,sol_CN[time-1],x,args=(alpha,),Left="Dirichlet",Right="Dirichlet")
    for space in range(1,len(x)-1,1):
        sol_exp[time,space] = alpha*sol_exp[time-1,space-1] + (1-2*alpha)*sol_exp[time-1,space]+ alpha*sol_exp[time-1,space+1]
    #sol_exp[time,0] = sol_exp[time,1]
    #sol_exp[time,len(x)-1] = sol_exp[time,len(x)-2]
```

```python
#print(sol)
ana = np.zeros((len(t),len(x)))
for time in range(0,len(t),1):
    ana[time] = np.sin(3*np.pi*x/(xmax-xmin))*np.exp(-D*9*np.pi**2*t[time]/((xmax-xmin)*(xmax-xmin)))
#print(ana)
```

```python
plt.plot(x,sol[-1,:],label='t=Implicit')
plt.plot(x,sol_CN[-1,:],label='t=last_time_CN')
plt.plot(x,sol_exp[-1,:],label='t=last_time_Explicit')
plt.plot(x,sol[0,:],label='t=0')
plt.plot(x,ana[-1,:],label='t=analytical')
plt.legend(loc='best')
plt.show()
```
