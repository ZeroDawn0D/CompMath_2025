```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#0 = > N (values)
#1 => C (values)
#2 => V (values)

#0 -> nothing
#1 -> N (border)
#2 -> C (border)
```

```python
exp = np.load('dataNCN/exp7.npy', allow_pickle=True).item()
grid = exp['NCV_grid']
```

```python
N_grid = grid[:,:,:, 0]
C_grid = grid[:,:,:,1]
V_grid = grid[:,:,:,2]
bool_grid = np.argmax(grid,-1)
bool_grid = (C_grid>(N_grid+V_grid)).astype(int)
```

```python
plt.imshow(bool_grid[990])
```
